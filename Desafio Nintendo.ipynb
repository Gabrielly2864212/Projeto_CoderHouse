```python
import pandas as pd

# 1 - Abrir o arquivo Excel em um DataFrame
caminho_arquivo = r'C:\Users\vinic\OneDrive\Área de Trabalho\Gabrielly\Curso Python - Coder House\Aulas e Exercícios\best-selling game consoles.xlsx'
df = pd.read_excel(caminho_arquivo, sheet_name='consoles')

# 2 - Substituir a palavra "NES" por "Nitendinho" no nome dos consoles e deixar todos os nomes em maiúsculas
df['Console Name'] = df['Console Name'].str.replace('NES', 'Nitendinho', case=False)
df['Console Name'] = df['Console Name'].str.upper()

# 3 - Filtrar os consoles com data de lançamento depois de 2010
df = df[df['Released Year'] > 2010]

# 4 - Descrever a base, exibir informações e substituir valores faltantes
df_describe = df.describe(include='all')  # Exibe as estatísticas descritivas
df_info = df.info()  # Exibe informações sobre o DataFrame
df.fillna('missing', inplace=True)  # Substitui valores ausentes por 'missing'

# 5 - Filtrar consoles descontinuados a menos de 2 anos da data de lançamento
df['Lifespan'] = df['Discontinuation Year'] - df['Released Year']
df_filtered = df[df['Lifespan'] < 2]

# Exibir o DataFrame final com os filtros aplicados
display(df_filtered)
```
